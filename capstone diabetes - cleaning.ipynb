```python
# checking for missing values
missing = pd.concat([X.isnull().sum(), X.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')

# no missing values, yay! :)
```

```python
# Smoker chart
smoke_value_counts = X['Smoker'].value_counts()

# Plot the bar chart using seaborn
sns.barplot(x=smoke_value_counts.index, y=smoke_value_counts.values)
```

```python
# Age chart
age_value_counts = X['Age'].value_counts()

# Plot the bar chart using seaborn
sns.barplot(x=age_value_counts.index, y=age_value_counts.values)
```

```python
# BMI chart

# X_reset = X.reset_index()
value_counts = X['BMI'].value_counts()

# Plot the bar chart using seaborn
sns.barplot(x=value_counts.index, y=value_counts.values)
plt.xticks(rotation=10)
plt.xticks(fontsize=6)
plt.figure(figsize=(100, 60))

step = 2  # Display every 2nd tick (adjust as needed)
plt.xticks(range(0, len(value_counts.index), step), value_counts.index[::step], rotation=45)

plt.show()
```
